# Doodle Zoo
## Doodle Recognition Training Notebook

This notebook is used to make training the model easier, allowing for step-by-step training and validation of the model.

## Step 1: Import the necessary libraries and helper classes

```python
from utils.neural.data import DataLoader, DataVisualizer
from utils.neural.net import Net
```

## Step 2: Initialize the dataloader and visualizer modules

```python
loader = DataLoader()
visualizer = DataVisualizer(loader)
```

### Step 2.1: Ensure the dataloader and visualizer modules are working by visualizing a sample of the data

```python
visualizer.show_random_train_batch()
```

## Step 3: Initialize the model

```python
net = Net(loader)
```

## Step 4: Train the model

```python
train = False  # Set to True to train the network - allows running the notebook with "Run All" without training

if train:
    net.train()
```

## Step 5: Load the model

```python
net.load()
```

## Step 6: Analyze the model

### Step 6.1: Run predictions on a batch from the test set

```python
images, labels = loader.get_test_batch()
predicted = net.evaluate(images)

visualizer.show_batch(images, labels, predicted)
```

### Step 6.2: Evaluate the accuracy of the model over the entire test set

```python
net.test_accuracy(as_plot=True)
net.test_accuracy(by_class=True, as_plot=True)
```

### Step 6.3: Visualize the features of the model

```python
net.visualize_features()
```
